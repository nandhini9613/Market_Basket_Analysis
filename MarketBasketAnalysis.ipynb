```python
# Convert 'Date' column to datetime objects
groceries['Date'] = pd.to_datetime(groceries['Date'], format='%d-%m-%Y')
```

```python
groceries.head()
```

```python
# Most Frequent Items (Bar)

top_items = groceries['itemDescription'].value_counts().head(10).reset_index()
top_items.columns = ['Item', 'Count']
fig1 = px.bar(top_items, x='Item', y='Count', title='Top 10 Most Purchased Items')
fig1.show()
```

```python
#Purchase Trend Over Time

daily_counts = groceries.groupby('Date').size().reset_index(name='Purchases')

fig2 = px.line(daily_counts, x='Date', y='Purchases', title='Purchase Trend Over Time')
fig2.show()
```

```python
#Treemap of Items
fig4 = px.treemap(top_items, path=['Item'], values='Count', title='Item Popularity Treemap')
fig4.show()
```

```python
#Unique Items Per Member (Bar)
unique_items = groceries.groupby('memberNumber')['itemDescription'].nunique().reset_index()
unique_items.columns = ['memberNumber', 'Unique_Items']

top_members = unique_items.sort_values('Unique_Items', ascending=False).head(10)

fig5 = px.bar(top_members, x='memberNumber', y='Unique_Items',
              title='Top 10 Members by Unique Items Bought')
fig5.show()
```
